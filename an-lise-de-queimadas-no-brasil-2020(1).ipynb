```python
req = requests.get(url_brasil_localidades)
```

```python
%%time

for arq in dados_ibge.find_all('a'):
    if 'BR_Localidades' in arq.get_text():
        print(arq.get_text())
        req = requests.get(url_brasil_localidades + arq["href"])
        print(req.status_code)
        open('/kaggle/working/mapas/' + arq['href'], 'wb').write(req.content)
```

### 3.2) Agregando dados de duas camadas

```python
agregados = gpd.sjoin(queimadas_risco_alto, brasil_malhas_municipios, op='within')
```

```python
agregados.head()
```

```python
queimadas_muni = agregados.groupby(by='NM_MUN').size()
queimadas_muni = queimadas_muni.sort_values(ascending=False)
queimadas_muni
```

### 3.3) Normalizando os dados: Exemplo com RR

```python
queimadas_muni_RR = agregados[agregados['SIGLA_UF'] == 'RR'].groupby(by='NM_MUN').size()
queimadas_muni_RR = queimadas_muni_RR.sort_values(ascending=False)
queimadas_muni_RR_df = queimadas_muni_RR.to_frame()
queimadas_muni_RR_df['MUN'] = queimadas_muni_RR_df.index.values
queimadas_muni_RR_df.columns = ['qtd_queimadas','MUN']
queimadas_muni_RR_df['AREA'] = 0
queimadas_muni_RR_df['NORM'] = 0
queimadas_muni_RR_df
```

```python
for mun in queimadas_muni_RR_df['MUN'].unique():
    print(mun)
    print(agregados[agregados['NM_MUN'] == mun]['AREA_KM2'].values[0])
    print(queimadas_muni_RR_df[queimadas_muni_RR_df['MUN'] == mun]['qtd_queimadas'].values[0])
    print('--------')
    
    queimadas_muni_RR_df.at[mun, 'AREA'] = agregados[agregados['NM_MUN'] == mun]['AREA_KM2'].values[0]
    queimadas_muni_RR_df.at[mun, 'NORM'] = agregados[agregados['NM_MUN'] == mun]['AREA_KM2'].values[0] / queimadas_muni_RR_df[queimadas_muni_RR_df['MUN'] == mun]['qtd_queimadas'].values[0]
```

### 4.2) Visualizando biomas em camadas

```python
agregados = gpd.sjoin(queimadas_risco_alto, brasil_malhas_municipios, op='within')
agregados.head()
```

```python
queimadas_muni = agregados.groupby(by=['NM_MUN','bioma']).size()
queimadas_muni = queimadas_muni.sort_values(ascending=False)
queimadas_muni
```

```python
queimadas_muni_df = queimadas_muni.to_frame()
queimadas_muni_df
```

```python
queimadas_muni_df = queimadas_muni_df.reset_index()
queimadas_muni_df.columns = ['NM_MUN', 'Bioma', 'qtd_queimadas']
queimadas_muni_df
```

```python
queimadas_muni_geom_df = pd.merge(brasil_malhas_municipios, queimadas_muni_df, on='NM_MUN')
queimadas_muni_geom_df
```

```python
coord_centro_brasil = [-14.235004, -51.925282]
mapa = folium.Map(location=coord_centro_brasil, zoom_start=4)
```

```python
folium.Choropleth(geo_data = queimadas_muni_geom_df[queimadas_muni_geom_df['Bioma'] == 'Pantanal'],
                  name = 'Bioma: Pantanal',
                  data = queimadas_muni_geom_df[queimadas_muni_geom_df['Bioma'] == 'Pantanal'],
                  columns = ['NM_MUN', 'qtd_queimadas'],
                  key_on = 'feature.properties.NM_MUN',
                  fill_color = 'YlGnBu',
                  fill_opacity = 0.75,
                  line_opacity = 0.5,
                  legend_name = 'Quantidade de queimadas por município (Pantanal)').add_to(mapa)

folium.Choropleth(geo_data = queimadas_muni_geom_df[queimadas_muni_geom_df['Bioma'] == 'Cerrado'],
                  name = 'Bioma: Cerrado',
                  data = queimadas_muni_geom_df[queimadas_muni_geom_df['Bioma'] == 'Cerrado'],
                  columns = ['NM_MUN', 'qtd_queimadas'],
                  key_on = 'feature.properties.NM_MUN',
                  fill_color = 'YlGnBu',
                  fill_opacity = 0.75,
                  line_opacity = 0.5,
                  legend_name = 'Quantidade de queimadas por município (Cerrado)').add_to(mapa)

folium.LayerControl().add_to(mapa)
```

```python
mapa.save("biomas_queimadas_mapa.html")
```
